```python
import pandas as pd
import numpy as np
import uuid
```

```python

df = pd.read_parquet("../data/nyc_yellow_taxi_dataset/yellow_tripdata_2024-01.parquet")
df.info()
```

```python
df.describe()
```

```python
df = df[
    [
        "tpep_pickup_datetime",
        "tpep_dropoff_datetime",
        "PULocationID",
        "DOLocationID",
        "passenger_count",
        "trip_distance",
        "fare_amount",
        "total_amount"
    ]
]
```

```python
df = df.dropna()

df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])

df = df[df["trip_distance"] > 0]
df = df[df["passenger_count"] > 0]
```

```python
df.info()
```

```python
# Trip duration in minutes
df["trip_duration_min"] = (
    (df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"])
    .dt.total_seconds() / 60
)

# Speed (km/h)
df["speed_kmh"] = df["trip_distance"] / (df["trip_duration_min"] / 60)
```

```python
df["trip_id"] = [str(uuid.uuid4()) for _ in range(len(df))]
```

```python
df = df.rename(columns={
    "tpep_pickup_datetime": "pickup_datetime",
    "tpep_dropoff_datetime": "dropoff_datetime",
    "PULocationID": "pickup_location_id",
    "DOLocationID": "dropoff_location_id"
})
```

```python
df.head()
```

```python
zone_df = pd.read_csv("../data/nyc_yellow_taxi_dataset/taxi_zone_lookup.csv")
```

```python
df = df.merge(
    zone_df.rename(columns={"LocationID": "pickup_location_id", "Borough": "pickup_borough"})[["pickup_location_id","pickup_borough"]],
    on="pickup_location_id",
    how="left"
)

df = df.merge(
    zone_df.rename(columns={"LocationID": "dropoff_location_id", "Borough": "dropoff_borough"})[["dropoff_location_id","dropoff_borough"]],
    on="dropoff_location_id",
    how="left"
)
```

```python
df["pickup_location"] = df.apply(
    lambda x: {
        "location_id": int(x["pickup_location_id"]),
        "borough": x["pickup_borough"]
    },
    axis=1
)

df["dropoff_location"] = df.apply(
    lambda x: {
        "location_id": int(x["dropoff_location_id"]),
        "borough": x["dropoff_borough"]
    },
    axis=1
)
```

```python
final_cols = [
    "trip_id",
    "pickup_datetime",
    "dropoff_datetime",
    "pickup_location",
    "dropoff_location",
    "passenger_count",
    "trip_distance",
    "fare_amount",
    "total_amount",
    "trip_duration_min",
    "speed_kmh"
]

df_final = df[final_cols]
```

```python
df_final.to_json("../data/cleaned_data/taxi_events.jsonl", orient="records", lines=True, date_format="iso")

print("✅ Data cleaned and exported to JSONL")
df_final.head()
```
